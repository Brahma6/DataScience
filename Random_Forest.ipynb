```python
data_aggrgated = pd.read_csv('save_perfect_data_12994.csv')
```

```python
print(data_aggrgated.shape)
data_aggrgated.head()
```

```python
data_y  = data_aggrgated['drivingstyle']
cols=['ID','drivingstyle','precd_veh_id']
data_x  = data_aggrgated.drop(cols,1)
```

```python
print(data_x.shape)
print(data_y.shape)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.cross_validation import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data_x, data_y, test_size = 0.05, random_state = 0)
```

```python
X, y = x_train, y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfclsifer = RandomForestClassifier(random_state = 0)
rfclsifer.fit(X, y)
```

```python
# Predicting the Test set results
y_pred = rfclsifer.predict(x_test)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)

print(cm)
```
